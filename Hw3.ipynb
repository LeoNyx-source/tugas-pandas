```python
import pandas as pd

file_path = "LB_TA.txt"

def calculate_average_at(file_path):
    #Membaca file dengan pemisah spasi/tab
    df = pd.read_csv(file_path, sep='\s+', engine='python')

    #mengecek kolom tic dan at di dalam file txt
    if 'tic' not in df.columns or 'at' not in df.columns:
        raise ValueError("Kolom 'tic' atau 'at' tidak ditemukan dalam file.")

    #Konversi kolom "at" ke numerik dan tandai NaN
    df['at'] = pd.to_numeric(df['at'], errors='coerce')

    #Identifikasi "tic" yang memiliki setidaknya satu nilai NaN di "at"
    invalid_tics = df[df['at'].isna()]['tic'].unique()

    #Membuang semua baris yang memiliki "tic" yang tidak lengkap
    df_clean = df[~df['tic'].isin(invalid_tics)]

    #Menghitung rata-rata "at" berdasarkan "tic"
    avg_at_per_tic = df_clean.groupby('tic', as_index=True)['at'].mean()

    return avg_at_per_tic

# Panggil fungsi dan cetak hasil
average_at = calculate_average_at(file_path)
print(average_at)
```
