```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("electronic-card-transactions-january-2025-csv-tables.csv")

# Pastikan "Group" ada dalam dataset
print("Available columns:", df.columns.tolist())

# Visualisasi jumlah transaksi berdasarkan kategori "Group"
plt.figure(figsize=(12, 6))
df['Group'].value_counts().nlargest(40).plot(kind='bar', color='skyblue')
plt.title("Number of Transactions by Group")
plt.ylabel('Number of Transactions')
plt.xlabel('Group')
plt.xticks(rotation=45)
plt.show()

# Mengatur ukuran figur untuk visualisasi berikutnya
plt.figure(figsize=(10, 5))

# Pilih hanya kolom numerik sebelum menghitung korelasi
df_numeric = df.select_dtypes(include=['number'])

# Hitung korelasi antar kolom numerik
c = df_numeric.corr()

# Plot heatmap korelasi
sns.heatmap(c, cmap="BrBG", annot=True)
plt.title("Correlation Heatmap")
plt.show()

# Scatter plot antara "Period" dan "Data_value" (bisa disesuaikan)
fig, ax = plt.subplots(figsize=(10,6))
ax.scatter(df['Period'], df['Data_value'], alpha=0.5, color='red')
ax.set_xlabel('Period')
ax.set_ylabel('Data Value')
plt.title("Scatter Plot of Period vs Data Value")
plt.show()
```
